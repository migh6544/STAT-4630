## A.1 Logistic regression log-likelihood

Consider the binomial regression model with the logit link function. That is:

1. Let $Y_i \sim binomial(n_i,p_i)$, be a binomial response (random component)


2. For the $i^{th}$ row of the covariate class, $x_{i,1},...,x_{i,p}$, $\eta_i = \beta_0 + \beta_1x_{i,1} + ... + \beta_px_{i,p}$ (systematic component)


3. Let $g(p_i) = \log\left(\frac{p_i}{1-p_i} \right)$ (link function)

**[10 points] Show that the log-likelihood function for binomial regression is given by**

\begin{align*}
\ell(\boldsymbol\beta) \propto \sum y_i\eta_i - \sum n_i\log\left(1 + e^{\eta_i} \right).
\end{align*}

First, note that $\eta_i \overset{set}{=} \log\left(\frac{p_i}{1-p_i} \right)$ implies that $p_i = \frac{e^{\eta_i}}{1+e^{\eta_i}}$.

The marginal pmf of the binomial is $f(y_i \, | \, n_i, p_i) \propto p_i^{y_i}(1-p_i)^{n_i-y_i}$ and thus the likelihood is

\begin{align*}
L(\boldsymbol\beta) = f(\mathbf{y} \, | \, \mathbf{n}, \mathbf{p} ) \propto \prod_{i=1}^np_i^{y_i}(1-p_i)^{n_i-y_i},
\end{align*}

and the log-likelihood is

\begin{align*}
\ell(\boldsymbol\beta) &\propto \sum_{i=1}^n\log\left( p_i^{y_i}(1-p_i)^{n_i-y_i} \right)\\
&\propto \sum_{i=1}^n{y_i}\log\left( p_i\right) + \sum_{i=1}^n(n_i-y_i) \log\left( 1-p_i \right) \\
&\propto\sum_{i=1}^n{y_i}\log\left( \frac{e^{\eta_i}}{1+e^{\eta_i}}\right) + \sum_{i=1}^n(n_i-y_i) \log\left( 1-\frac{e^{\eta_i}}{1+e^{\eta_i}} \right) \\
&\propto\sum_{i=1}^n{y_i}\log\left( \frac{e^{\eta_i}}{1+e^{\eta_i}}\right) + \sum_{i=1}^n(n_i-y_i) \log\left(\frac{1}{1+e^{\eta_i}} \right) \\
&\propto\sum_{i=1}^n{y_i}{\eta_i} - \sum_{i=1}^n{y_i}\log\left(1+e^{\eta_i}\right) + \sum_{i=1}^n(n_i-y_i) \left[  \underbrace{\log\left(1 \right) }_{0} - \log\left(1+e^{\eta_i} \right) \right] \\
&\propto\sum_{i=1}^n{y_i}{\eta_i} - \sum_{i=1}^n{y_i}\log\left(1+e^{\eta_i}\right) -\sum_{i=1}^nn_i \log\left(1+e^{\eta_i} \right)  + \sum_{i=1}^n y_i  \log\left(1+e^{\eta_i} \right)  \\
&\propto\sum_{i=1}^n{y_i}{\eta_i} -\sum_{i=1}^nn_i \log\left(1+e^{\eta_i} \right)    \\
\end{align*}

## A. 2 Poisson regression rate model log-likelihood

Consider the Poisson regression model with the log link function. That is:

1. Let $Y_i \sim Poisson(\lambda_i)$, be a Poisson response (random component).


2. For the $i^{th}$ row of the covariate class, $x_{i,1},...,x_{i,p}$, $\eta_i = \beta_0 + \beta_1x_{i,1} + ... + \beta_px_{i,p}$ (systematic component).



3. Let $o_i$ be the $i^{th}$ measurement of the "exposure" variable (e.g., period of time, area in space).


4. Let $g\left(\frac{\lambda_i}{o_i}\right) = \log\left(\frac{\lambda_i}{o_i} \right)$ (link function).
 

**[10 points] Show that the log-likelihood function for Poisson regression with an offset is is given by**


\begin{align*}
\ell(\boldsymbol\beta) \propto \sum y_i\eta_i - \sum o_ie^{\eta_i}  + \sum y_i\log(o_i).
\end{align*}

First, note that $\eta_i \overset{set}{=} \log\left(\frac{\lambda_i}{o_i}\right) $ implies that $\lambda_i = \log{\eta_i} + \log(o_i)$.

The marginal pmf of the binomial is $f(y_i \, | \,\lambda_i) \propto e^{-\lambda_i}\lambda_i^{y_i}$ and thus the likelihood is

\begin{align*}
L(\boldsymbol\beta) = f(\mathbf{y} \, | \,\boldsymbol\lambda) = \prod_{i=1}^n f(y_i \, | \,\lambda_i) = \prod_{i=1}^n e^{-\lambda_i}\lambda_i^{y_i},
\end{align*}

and the log-likelihood is

\begin{align*}
\ell(\boldsymbol\beta) =   \sum_{i=1}^n\log  f(y_i \, | \,\lambda_i) &\propto \sum_{i=1}^n \log \left(e^{-\lambda_i}\lambda_i^{y_i}\right)\\
&\propto\sum_{i=1}^n -\lambda_i  + \sum_{i=1}^n {y_i}\log\lambda_i\\
&\propto   \sum_{i=1}^n {y_i}\left(\eta_i + \log o_i \right) - \sum_{i=1}^n e^{\eta_i}o_i\\
&\propto   \sum_{i=1}^n {y_i}\eta_i + \sum_{i=1}^n y_i\log o_i  - \sum_{i=1}^n e^{\eta_i}o_i\\
\end{align*}

### A.3 Justifying Metropolis-Hastings 

The proof that the simulated sequence of Metropolis-Hastings iterations $\theta^{(1)}, \theta^{(2)}, . . .$ converges to the target distribution has two steps: 

1. Show that the simulated sequence is a Markov chain with a unique stationary distribution, 


2. Show that the stationary distribution equals the target (posterior) distribution, $\pi(\theta \, | \, x)$.

The first step of the proof holds if the Markov chain is *irreducible*, *aperiodic*, and not *transient*. 

**A.3 (a) [8 points] Define the terms *irreducible*, *aperiodic*, and *transient* in the context of Markov chain theory, and describe why the simulated sequence satisfies these properties.**

You do not need a formal proof. A description will suffice.

For definitions, see [here](https://similarweb.engineering/mcmc/).

Except for trivial exceptions, the latter two conditions hold for a random walk on any proper distribution, and irreducibility holds as long as the random walk has a positive probability of eventually reaching any state from any other state; that is, the jumping distributions Jt must eventually be able to jump to all states with positive probability.

**A.3 (b) (STAT 5630 Only, 10 points) Show that the stationary distribution for the Metropolis- Hastings algorithm is, in fact, the target distribution, $\pi(\theta\, | \, x)$.**

To see that the target distribution is the stationary distribution of the Markov chain generated by the Metropolis-Hastings algorithm, consider starting the algorithm at time $t − 1$ with a draw $\theta^{(t−1)}$ from the target distribution $\pi(\theta \, | \, x)$. Now consider any two such points $\theta_a$ and $\theta_b$, drawn from $\pi(\theta \, | \, x)$ and labeled so that $\pi(\theta_b \, | \, x) \ge \pi(\theta_a \, | \, x)$. The unconditional probability density of a transition from $\theta_b$ to $\theta_a$ is

$$\pi(\theta^{(t−1)} = \theta_a, \theta^{(t)} = \theta_b) = \pi(\theta_a\, | \, x)J_t(\theta_b \, | \, \theta_a),$$

where the acceptance probability is 1 because of our labeling of $a$ and $b$, and the unconditional probability density of a transition from $\theta_a$ to $\theta_b$ is

$$\pi(\theta^{(t)} = \theta_a, \theta^{(t−1)} = \theta_b) = \pi(\theta_b \, | \, x)J_t(\theta_a \, | \, \theta_b)\left( \frac{ \pi(\theta_a \, | \, x)J_t(\theta_b \, | \, \theta_a) }{\pi(\theta_b \, | \, x)J_t(\theta_a \, | \, \theta_b)} \right) = \pi(\theta_a \, | \, x)J_t(\theta_b \, | \, \theta_a),$$
  
Since their joint distribution is symmetric, $\theta^{(t)}$ and $\theta^{(t−1)}$ have the same marginal distributions, and so $\pi(\theta \, | \, x)$ is the stationary distribution of the Markov chain of $\theta$.

Consider `x` in the cell below to be data. Let's suppose that a group of two researchers, neither of which knows the true mean `mu_true`, wishs to estimate it with a Bayesian poterior. However, there is disagreement among the researchers' priors:

- Researcher #1: $\mu \sim N(0,1)$

- Researcher #2: $\mu \sim N(50,1)$

Further, Researcher #1 and Researcher #2 have roughly equal expertise about the data generating process, with researcher #1 being slightly more knowledeable. More precisely, we weight each researchers' prior according to `alpha = c(0.6,0.4)`.


**B.1 (a) [5 points] Construct a prior distribution as a mixture of each individual researcher's prior. Plot the prior.**

```r
set.seed(5630)
n = 7; mu_true = 45; sig_true = 35; x = rnorm(n,mu_true,sig_true)
```

```r
alpha = c(0.6, 0.4); 
mu0 = c(0,50)
m = 100; mu_grid = seq(-10,55,length.out = m)

prior = function(mu,alpha){
    pi_mu = 0
    for (i in 1:length(alpha)){
    pi_mu = pi_mu + alpha[i]*dnorm(mu,mu0[i],1) 
}
    return(pi_mu)
}


plot(mu_grid, prior(mu_grid,alpha),type = "l")
```

**B.1 (b) [3 points] Construct and plot the likelihood function for the data.**

```r
l = function(mu,x){
    l = 1
    for (i in 1:length(x)){
        l = l*dnorm(x[i],mu,sig_true)
    }
    return(l)
}
plot(mu_grid, l(mu_grid,x), type = "l")
```

**B.1 (c) [10 points] Construct and plot the posterior up to a normalizing constant. Describe it's form.**

```r
post = function(mu,alpha,x){
    post = l(mu,x)*prior(mu,alpha)
    return(post)
}

logpost = function(mu,alpha,x){
    logpost = log(l(mu,x)) + log(prior(mu,alpha))
    return(logpost)
}

logpost(1,alpha,x)

plot(mu_grid, post(mu_grid,alpha,x), type = "l", xlab = expression(mu), ylab = expression(posterior))
```

There are clearly two modes to this posterior.

**Produce a trace plot and comment on the convergence of the chain.**

```r
mhs = function(mu0,proposal_sd){
    samples = matrix(NA,T,length(mu0))
    mu = mu0
    for (t in 1:T){  
        proposal = rnorm(1,mu,proposal_sd); #sample a proposal
        logR = logpost(proposal,alpha,x) - 
            logpost(mu,alpha,x); #compute the log of the ratio
        if(log(runif(1))<logR){
            mu = proposal #accept/reject the proposal
            }
            samples[t] = mu #store the accept/reject in the chain
            }
    return(samples)
}

T = 50000
chain1 = matrix(NA,T,1); 
chain1 = mhs(5,1)
#chain2 = mhs2(mu)
```
