### A.2 MVN conditional expectation and variance

Let $\mathbf{Y} = (Y_1,Y_2)^T$ be a multivariate normal vector:

$$\mathbf{Y} \sim N\left(\boldsymbol{\mu}, \Sigma \right),$$ 

where $\boldsymbol{\mu} = (\underbrace{ \mu_1 }_{E(Y_1)},\underbrace{ \mu_2 }_{E(Y_2)})^T$ is a vector of means and 

$\Sigma = \left(\begin{matrix}
\sigma_{1,1} & \sigma_{1, 2} \\
\sigma_{2, 1} & \sigma_{2,2} \\
\end{matrix}\right)$

and $\Sigma$ is a *variance-covariance* matrix with entries $\displaystyle \sigma_{i, j} = Cov(Y_i,Y_j)$ for $i,j = 1,2$ (specifically, this implies that $\sigma_{i,i} = Var(Y_i)$). It turns out that the conditional distribution $Y_1 \, | \, Y_2 = y_2$ is normal. In this problem, we will find the mean and variance of this normal distribution.


**A.2 (a) [5 points]  Define $V = Y_1 + aY_2$ where $a = -\frac{\sigma_{1, 2}}{\sigma_{2,2}}$. Show that $V$ and $Y_2$ are uncorrelated, i.e., that $Cov(V, Y_2) = 0$. (Since both $V$ and $Y_2$ are normal, it follows that they will be independent.)**

\begin{align*}
Cov(V, Y_2) &= Cov(Y_1 + aY_2, Y_2) = Cov(Y_1, Y_2) + aCov(Y_2, Y_2) \\
&= \sigma_{1,2} - \frac{\sigma_{1, 2}}{\sigma_{2,2}}\sigma_{2,2} = \sigma_{1, 2} - \sigma_{1, 2} = 0
\end{align*}

**A.2 (b) [7 points] Use $V$ to find $E(Y_1 \, | \, Y_2 = y_2)$.**

$V = Y_1 + aY_2$ implies that $Y_1 = V - aY_2$. Thus,

\begin{align*}
E(Y_1 \, | \, Y_2 = y_2) &= E(V - aY_2 \, | \, Y_2 = y_2) = E(V \, | \, Y_2 = y_2) - ay_2 
\end{align*}

Since $V$ is independent of $Y_2$, $E(V \, | \, Y_2 = y_2) \overset{i}{=} E(V) = E(Y_1 + aY_2) = \mu_1 + a\mu_2$. Thus,


\begin{align*}
E(Y_1 \, | \, Y_2 = y_2)  &=\mu_1 + a\mu_2 - ay_2 = \mu_1 + a(\mu_2 - y_2) = \mu_1 -\frac{\sigma_{1, 2}}{\sigma_{2,2}}(\mu_2 - y_2) 
\end{align*}

**A.2 (c) [9 points]  Use $V$ to find $Var(Y_1 \, | \, Y_2 = y_2)$.**

\begin{align*}
Var(Y_1 \, | \, Y_2 = y_2) &= Var(V - aY_2 \, | \, Y_2 = y_2) = Var(V \, | \, Y_2 = y_2) + Var(aY_2 \, | \, Y_2 = y_2) - 2a\underbrace{Cov(V, Y_2 \, | \, Y_2 = y_2)}_{0}\\
&= Var(V \, | \, Y_2 = y_2) + a^2\underbrace{Var(Y_2 \, | \, Y_2 = y_2)}_{0} = Var(V \, | \, Y_2 = y_2) \overset{i}{=} Var(V) \\
&= Var(Y_1 + aY_2) = Var(Y_1) + Var(aY_2) + 2aCov(Y_1, Y_2) \\
&= \sigma_{1,1} + \left(\frac{\sigma_{1, 2}}{\sigma_{2,2}}\right)^2\sigma_{2,2} -2\frac{\sigma_{1, 2}}{\sigma_{2,2}}\sigma_{1,2} = \sigma_{1,1} + \frac{\sigma_{1, 2}^2}{\sigma_{2,2}} -2\frac{\sigma_{1, 2}^2}{\sigma_{2,2}} \\
&= \sigma_{1,1} -\frac{\sigma_{1, 2}^2}{\sigma_{2,2}}
\end{align*}


**A.2 (d) [4 points] Let $\rho = \frac{\sigma_{1,2}}{\sqrt{\sigma_{1,1}\sigma_{2,2}}}$. Show that $Var(Y_1 \, | \, Y_2 = y_2) = (1-\rho^2)\sigma_{1,1}$.**

If $\rho^2 = \frac{\sigma_{1,2}^2}{\sigma_{1,1}\sigma_{2,2}}$ then $\sigma_{1,1}\rho^2 = \frac{\sigma_{1,2}^2}{\sigma_{2,2}}$ and $Var(Y_1 \, | \, Y_2 = y_2)$ becomes:

\begin{align*}
Var(Y_1 \, | \, Y_2 = y_2) &= \sigma_{1,1} -\sigma_{1,1}\rho^2 = (1-\rho^2)\sigma_{1,1}
\end{align*}

### A.3 Bivariate normal: inference on the correlation coefficient

Assume that $(X,Y)$ follow a bivariate normal distribution with $E(X) = E(Y) = 0$ and $Var(X) = Var(Y) = 1$. The pdf of $(X,Y)$ is given as

\begin{align*}
f(x,y \, | \, \rho) = \frac{1}{2\pi\sqrt{1-\rho^2}}\exp{\left\{-\frac{x^2 - 2\rho xy + y^2}{2(1-\rho^2)} \right\} },
\end{align*}

where $\rho$ is the correlation coefficient for $X$ and $Y$. Suppose that we observe $n$ iid realizations of $(X,Y)$. That is, we observe:

| X | Y |
|---|---|
| $x_1$ | $y_1$ |
| $x_2$ | $y_2$ |
| $\vdots$ | $\vdots$ |
| $x_n$ | $y_n$ |

**A.3 (a) [5 points] Find the joint pdf and the likelihood function for these observations.**

The joint pdf is:

\begin{align*}
f(\mathbf{x},\mathbf{y} \, | \, \rho) &= \prod^n_{i=1} \left(\frac{1}{2\pi\sqrt{1-\rho^2}}\exp{\left\{-\frac{x_i^2 - 2\rho x_iy_i + y_i^2}{2(1-\rho^2)} \right\} } \right) \\
&= \left(\frac{1}{2\pi\sqrt{1-\rho^2}}\right)^{n}\exp{\left\{-\frac{\sum^n_{i=1}x_i^2 - 2\rho \sum^n_{i=1}x_iy_i + \sum^n_{i=1}y_i^2}{2(1-\rho^2)} \right\} }
\end{align*}

The likelihood is this same functional form, but with $\rho$ as the variable and $x$ and $y$ fixed.

**A.3 (b) [5 points] Assume that $\rho \sim U(-1,1)$. Up to a constant with respect to $\rho$, write down the posterior distribution for $\rho$ given the data $(x_i, y_i)$, $i = 1,...,n$.**

HINT: You do not have to compute the integral denominator of Bayes' theorem here!

\begin{align*}
\pi(\rho \, | \, x,y) &= \frac{f(\mathbf{x},\mathbf{y} \, | \, \rho) \pi(\rho)}{f(x,y)} \\
&\propto \left(\frac{1}{2\pi\sqrt{1-\rho^2}}\right)^{n}\exp{\left\{-\frac{\sum^n_{i=1}x_i^2 - 2\rho \sum^n_{i=1}x_iy_i + \sum^n_{i=1}y_i^2}{2(1-\rho^2)} \right\} }\left(\frac{1}{2}\right)
\end{align*}

## B.1 Poisson inference

Suppose that the number of species discovered each year is thought to be Poission distributed with rate parameter $\lambda$; that is, for year $t$, $X_t \, | \, \lambda \overset{iid}{\sim} Poisson(\lambda)$. We can use Bayesian inference with real data, $\mathbf{x}$, to estimate the true $\lambda$. Suppose that our prior distribution on $\lambda$ is uniform between $0$ and $100$; that is, $\lambda \sim U(0,100)$. Further, suppose that, from 2010 to 2015, the number of new species discovered per year (in ascending order) is recorded as $\mathbf{x} = (64,13,33,18,30,20)$. Our goal in this problem will be to plot the posterior distribution of $\lambda \, | \, \mathbf{x}$.

**B.1(a) [3 points] First, construct a grid of $m =500$ evenly spaced values for $\lambda$ between $0$ and $100$. The function `seq()` may help here. Store these values in a vector `lambda`. Then, compute the prior density at each value of $\lambda$, (denoated $\pi(\lambda)$ below).**

```r
lambda = seq(0,100,length.out = 500); length(lambda)
prior = dunif(lambda, 0, 100)
```

**B.1(b) [5 points] Then, construct the likelihood function for `x = c(64,13,33,18,30,20)` (denoted $f(\mathbf{x} \, | \, \lambda)$ below). Remember, since our data are iid, the likelihood function is the product of the marginal pmfs, with the data fixed at $x$, and with $\lambda$ as a variable.**

```r
x = c(64,13,33,18,30,20)
likelihood = 1
for(t in 1:6){likelihood=likelihood*dpois(x[t],lambda)}
```

**B.1(c) [6 points] Compute and plot the posterior distribution of $\lambda \, | \, \mathbf{x}$. Comment on the shape of the distribution.**

Here, the posterior distribution, denoted $\pi(\lambda \, | \, \mathbf{x} )$, is given by

\begin{align*}
\pi(\lambda \, | \, \mathbf{x} ) = \frac{f(\mathbf{x} \, | \, \lambda )\pi(\lambda) }{\sum^{500}_{i = 1}f(\mathbf{x} \, | \, \lambda_i )\pi(\lambda_i)},
\end{align*}

where $\displaystyle\sum^{500}_{i = 1}f(\mathbf{x} \, | \, \lambda_i )\pi(\lambda_i)$ is the sum of the likelihood times the prior over all values of $\lambda$ in `lambda` from part B.1(a).

```r
posterior = likelihood*prior/sum(likelihood*prior); sum(posterior)
plot(lambda, posterior, type = "l")
```

It looks relatively normal, centered around $30$.

**B.2 (a) [4 points] Plot the prior distribution for $n$. Choose values of $n$ between $0$ and $30$.**

**B.2 (b) [6 points] Assuming $\theta = 0.4$ is known, and that there were $y = 0$ purchases on the day of the sale, compute the posterior distribution for $n \, | \, y$. Plot the distribution. How does it compare to the prior distribution?**

```r
#I'm doing work here for the next few parts of this problem

theta = c(0.4, 0.8);
y = c(0,5,10);
n = 0:30
prior = dpois(n,10)

#theta = 0.4
l = cbind(dbinom(y[1],n,theta[1]),dbinom(y[2],n,theta[1]),dbinom(y[3],n,theta[1]))
posterior_theta1 = cbind(prior*l[,1]/(sum(prior*l[,1])),
                  prior*l[,2]/(sum(prior*l[,2])),prior*l[,3]/(sum(prior*l[,3])))
colSums(posterior_theta1) #check that the posterior distributions sum to one...
```
