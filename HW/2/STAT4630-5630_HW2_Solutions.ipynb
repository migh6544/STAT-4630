## A.1 [8 points] Gamma/Exponential

**Let $X_1,...,X_n \overset{iid}{\sim} Exp(\lambda)$, and let the prior distribution on $\lambda$ be given by $\lambda \sim \Gamma(\alpha, \beta)$. Find the posterior distribution of $\lambda$ given the data.**

Note: 

The convention in this course will be to interpret $\Gamma(\alpha, \beta)$ as the "shape/rate" [parameterization](https://en.wikipedia.org/wiki/Gamma_distribution): If $Y \sim \Gamma(\alpha, \beta)$ then the pdf of $Y$ is given by 

\begin{align*}
f(y \, | \, \alpha, \beta) = \frac{\beta^\alpha}{\Gamma(\alpha)}y^{\alpha -1}e^{-\beta y}, \,\,\,\,\, y >0, \,\, \alpha > 0, \,\, \beta >0.
\end{align*}

Above, $\Gamma(\alpha)$ is the [Gamma function](https://en.wikipedia.org/wiki/Gamma_function) of $\alpha$. In the Gamma distribution, $\Gamma(\alpha)$ is just a normalizing constant, and so you do not need to compute it explicitly (just leave it as $\Gamma(\alpha)$).

Note that the prior distribution function is given by 

\begin{align*}
\pi(\lambda) =  \frac{\beta^\alpha}{\Gamma(\alpha)}\lambda^{\alpha -1}e^{-\beta \lambda}, \,\,\,\,\, \lambda >0, \,\, \alpha > 0, \,\, \beta >0,
\end{align*}

and the likliehood is given by

\begin{align*}
f(\mathbf{x} \, | \, \alpha, \beta) = \lambda^ne^{-\lambda n \bar{x}}
\end{align*}

Thus, for prior $\pi(\lambda)$, the posterior is given, up to a normalizing constant, by:

\begin{align*}
\pi(\lambda \, | \, \mathbf{x}) &\propto f(\mathbf{x} \, | \, \alpha, \beta)\pi(\lambda) = \left(\lambda^ne^{-\lambda n \bar{x}}\right)\left(\frac{\beta^\alpha}{\Gamma(\alpha)}\lambda^{\alpha -1}e^{-\beta \lambda}\right) \\
&\propto \frac{\beta^\alpha}{\Gamma(\alpha)}\lambda^{n + \alpha -1}e^{-(n\bar{x} + \beta)\lambda}
\end{align*}

Up to a normalizing constant, this is a Gamma distribution. Thus, $\lambda \, | \, \mathbf{x} \sim \Gamma(n + \alpha, \, n\bar{x} + \beta)$

**A.2 (a) [5 points] How might Bayes' theorem help to distinguish between claims that companies are making? For example, can Bayes' theorem be used to show that a screening test may be *accurate* - i.e., show a postive result when the disorder is present - but also that a person taking the test shouldn't have "total confidence" in the results when they take it?**

Let $C = $ "the individual has the condition" and $+ = $ "the individual tests positive for the condition". Then $P(+ \, | \, C)$ is the probability of a positive result when the condition is present. $P(C \, | \, +)$ represents the probability that the individual has the condition, given a positive result, i.e., the accuracy. This represents our "confidence" in the results, i.e., a high value for $P(C \, | \, +)$ suggests that we should believe with a high degree of confidence that the condition is present. Bayes' theorem tells us that

\begin{align*}
P(C \, | \, +) =  \frac{P(+ \, | \, C)P(C)}{P(+)}.
\end{align*}

Clearly, we can have a highly accurate test give a positive result and still have low confidence in the results. This will occur when the condition is rare, i.e., $P(C)$ is very low. This is exactly the case here, since companies are testing for rare conditions.

### A.3 Negative binomial conjugate prior

**A.3 [6 points] Let $X$ represent the number of failures observed before $r$ successes are observed in a set of independent Bernoulli trials. That is, let $X \sim NB(r, p)$ with $r$ known and $p$ unknown. In the first classwork assignment (the problem on stopping rules), we saw that, if $p \sim beta(\alpha, \beta)$, then the posterior distribution for $p \, | \, x$ is $beta(\alpha + r, \beta + x)$. Prove it!**

In this case, the pdf for $p$ is $\pi(p \, | \, \alpha, \beta) = \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha)\Gamma(\beta)}p^{\alpha-1}(1-p)^{\beta-1}$ and the likelihood is $f(x  \, | \, r, p) = {x + r -1 \choose x} p^r(1-p)^x$. This, the posterior is:

\begin{align*}
\pi(p \, | \, x) &\propto f(x  \, | \, r, p)\pi(p \, | \, \alpha, \beta) = p^r(1-p)^xp^{\alpha-1}(1-p)^{\beta-1} \\
&\propto p^{\alpha + r - 1}(1-p)^{\beta + x -1}.
\end{align*}

This result is the beta distribution with $\alpha_1 = \alpha + r$ and $\beta_1 = \beta + x$, up to the normalizing constant.

## B.1 Normal conjugate example

Suppose Company $A$ is considering using recycled cobalt in the smartphone batteries that it produces. Denote the true mean battery life (run time on a full charge) as $\mu$.  Company $A$ would like to know if there is evidence to strongly suggest that $\mu$ exceeds or falls short of $24$ hours ($24$ hours is the true mean battery life of the battery currently in use). Researchers' prior beliefs about $\mu$ can be represented by a normal distribution, centered at $\theta = 24$, with variance $\tau^2 = 4$. That is: $\mu \sim N(24, 4)$.

Suppose the researchers at company $A$ test $n = 25$ randomly selected phones with the recycled cobalt. Denote the battery life of phones in a possible sample as $X_1,...,X_n$. Assume that the standard deviation of battery life times is $\sigma = 1$) For this particular sample, the data $x_1,...,x_n$ were measured as:

\begin{align*}
&25.0, \, 23.8,\, 23.3, \,24.0, \,24.5, \,25.3, \,22.4, \,25.1, \,20.3, \,22.3, \,25.4, \,22.9, \,24.4, \,24.9, \,24.3, \\
&23.0, \,24.7, \,24.7, \,23.5, \,23.9, \,24.7, \,25.4, \,23.9, \,23.1, \,23.0. \\
\end{align*}


```r
x = c(25.0, 23.8, 23.3, 24.0, 24.5, 25.3, 22.4, 25.1, 20.3, 22.3, 25.4, 22.9, 24.4, 24.9, 24.3, 23.0, 24.7, 24.7, 23.5, 23.9, 24.7, 25.4, 23.9, 23.1, 23.0)
mean(x)
hist(x)
```

**B.1 (a) [6 points] Compute the posterior distribution mean and variance of $\mu \, | \, \mathbf{x}$. What would be a reasonable point estimate for $\mu$ given the data?**

```r
theta = 24; tau = 2; n = 25; sigma = 1; m = 0.25; sigma^1/m
xbar = mean(x)
post_mu = n*xbar/(n+m) + m*theta/(n+m)
#post_mu = (n*xbar*sigma^(-2) + theta*tau^(-2))/(n*sigma^(-2) + tau^(-2))

post_sig2 = sigma^2/(n+m)
#post_sig2 = 1/(n/sigma^2 + 1/tau^2)

post_mu; post_sig2;
```

**B.1 (b) [8 points] Plot the prior distribution and posterior distribution (on the same plot). What do you notice about this plot?**

```r
library(ggplot2)
mu_grid = seq(21,26, length.out = 300)

df = data.frame(mu = mu_grid, y_prior = dnorm(mu_grid, theta, tau), 
                y_post = dnorm(mu_grid, post_mu, sqrt(post_sig2)))

ggplot(df) +
    geom_line(aes(x = mu, y = y_prior), col = "#CFB87C") + 
    geom_line(aes(x = mu, y = y_post)) +
    xlab(expression(mu)) +
    ylab("Density") +
    theme_bw()
```

The prior reflects a high degree of uncertainty in the value of $\mu$. The posterior reflects much less uncertainty. 

**B.1 (c) [4 points] Compute a $95\%$ credible interval for $\mu$ given the data. Interpret this interval. In particular, do we have evidence to suggest that $\mu = 24$?**

```r
qnorm(c(0.025, 0.975), post_mu, sqrt(post_sig2))
```

Given the data, and the prior, probability that $\mu$ falls within $\approx (23.1, 23.9)$ is $0.95$. Yes, we do have evidence that $\mu = 24$.

**B.1 (d) [8 points] Now, suppose instead that prior to the experiment, researchers believed that $\theta = 13$ was the center of the prior distribution, with $\tau^2 = 1$. Compute the posterior distribution for this prior (for the same data as above), reproduce the plots from above, and describe the difference between this posterior and the one from previous parts.**

```r
theta = 13; tau = 1; n = 25; sigma = 1; m = 1
xbar = mean(x)
post_mu = n*xbar/(n+m) + m*theta/(n+m)
#post_mu = (n*xbar*sigma^(-2) + theta*tau^(-2))/(n*sigma^(-2) + tau^(-2))

post_sig2 = sigma^2/(n+m)
#post_sig2 = 1/(n/sigma^2 + 1/tau^2)

post_mu; post_sig2

mu_grid = seq(10,25, length.out = 300)

df = data.frame(mu = mu_grid, y_prior = dnorm(mu_grid, theta, tau), 
                y_post = dnorm(mu_grid, post_mu, sqrt(post_sig2)))

ggplot(df) +
    geom_line(aes(x = mu, y = y_prior), col = "#CFB87C") + 
    geom_line(aes(x = mu, y = y_post)) +
    xlab(expression(mu)) +
    ylab("Density") +
    xlim(c(10,25)) +
    theme_bw()
```

The prior has caused the center of the posterior to shift downward a bit.

**B.1 (e) [4 points] Compute a credible interval for this new posterior distribution. Interpret this interval. In particular, do we have evidence to suggest that $\mu = 24$?**

```r
qnorm(c(0.025, 0.975), post_mu, sqrt(post_sig2))
```

Given the data, and the prior, probability that $\mu$ falls within $\approx (23.1, 23.9)$ is $0.95$. No, we don't have strong evidence that $\mu = 24$.

**B.2 (c) [21 points] Suppose we'd like to make a prediction about the number of failures that we are likely to observe before the *next* $r = 3$ trials. Estimate the posterior predictive distribution using the algorithm described on page 31 of our textbook (*Bayesian Statistical Methods*). Plot this distribution.**

(There are three cells below to break up your code a bit. In the first, simulate the prediction values. In the second, calculate the relative frequences. In the third, plot the distribution.)

```r
set.seed(99)
m = 5000; pred_sim = matrix(NA, nrow = m) #initializing vector of simulated predictions

for (i in 1:m){
    p_star = rbeta(1, alpha + r, beta + x); #draw a posterior sample
    pred_sim[i] = rnbinom(1, 3, p_star) #prediction from the likelihood
}

pred_sim = pred_sim[order(pred_sim)]; #order vector of simulated predictions
```

```r
M = max(pred_sim) #largest # of failures

counts = matrix(0,nrow = M+1) #initializing vector to store counts 
for (j in 1:(M+1)){
    for (i in 1:m){
       counts[j] = counts[j] + ifelse(pred_sim[i] == j-1,1,0) #counter for each value of the simulated prediction
    }
}

ppd_sim = (counts/m) #relative frequency of simulated predictions
```

```r
library(ggplot2)
df = data.frame(x_star = 0:M, ppd_sim);
ggplot(data=df, aes(x=x_star, y=ppd_sim)) +
  geom_bar(stat="identity") + 
    ylim(c(0,0.15)) + 
    xlab("x") +
    ylab("Posterior predictive distribution") +
    theme_bw()
```
